## Dealing with Missing Sleep Timestamps
**Goal:** Fill missing sleep timestamps with 0 for every minute
1. add a full timeline 
2. fill values with zeros

```python
import pandas as pd

#import sleep csv
sleep = pd.read_csv("../src/data/user_5577150313_minuteSleep_merged.csv", parse_dates=['ActivityDateTime'])


start_time = sleep['ActivityDateTime'].min()
end_time = sleep['ActivityDateTime'].max()

timeline30sec = pd.date_range(start=start_time, end=end_time, freq='30s')
timeline30sec = pd.DataFrame(timeline30sec, columns=['ActivityDateTime'])

# Merge the 30-second timeline with the sleep data
full_sleep_30sec = timeline30sec.merge(sleep, how="left", on="ActivityDateTime")
full_sleep_30sec.fillna({'Id':5577150313}, inplace=True)
full_sleep_30sec.fillna({'value':0}, inplace=True)
full_sleep_30sec['Id'] = full_sleep_30sec['Id'].astype('int64')
full_sleep_30sec['value'] = full_sleep_30sec['value'].astype('int64')

# Now, we need to aggregate every two 30-second intervals into one minute
full_sleep_30sec['ActivityDateTime'] = pd.to_datetime(full_sleep_30sec['ActivityDateTime'])
full_sleep = (full_sleep_30sec.resample('1min', on='ActivityDateTime')
              .agg({'Id': 'first', 'value': 'sum'})
              .reset_index())

#test
# print(full_sleep['value'].sum())
# print(sleep['value'].sum())

# df = pd.read_csv('../src/data/user_5577150313_minuteSleep_preprocessed.csv', parse_dates=['ActivityDateTime'])
# print(df['value'].sum())
```
