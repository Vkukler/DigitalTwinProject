```python
import pandas as pd


df = pd.read_csv("../src/data/user_5577150313_minuteSleep_preprocessed.csv", parse_dates=['ActivityDateTime'])


daily_sleep = df[df['value'] != 0].groupby(df[df['value'] != 0]['ActivityDateTime'].dt.date)['value'].count()

for i in daily_sleep:
    print(f"{i//60} hours {i%60} minutes")
```

```python
import pandas as pd


steps = pd.read_csv(
    "../src/data/user_5577150313_minuteStepsNarrow_merged.csv",
    parse_dates=['ActivityMinute']
)

# Filter: keep only rows with steps > 0
filtered_steps = steps[steps['Steps'] > 0].copy()

# Group by date and sum total steps
daily_steps = (
    filtered_steps
    .groupby(filtered_steps['ActivityMinute'].dt.date)['Steps']
    .sum()
    .reset_index()
)

# Rename columns for clarity
daily_steps.columns = ['Date', 'TotalSteps']

# Print daily totals nicely formatted
print("Daily step totals:")
for _, row in daily_steps.iterrows():
    print(f"{row['Date']}: {row['TotalSteps']:,} steps")
```

```python
import pandas as pd


calories = pd.read_csv("../src/data/user_5577150313_minuteCaloriesNarrow_merged.csv", parse_dates=['ActivityMinute'])


daily_calories = calories.groupby(calories['ActivityMinute'].dt.date)['Calories'].sum()

# Optionally rename columns
daily_calories.columns = ['Date', 'TotalCalories']

print(daily_calories)
```
