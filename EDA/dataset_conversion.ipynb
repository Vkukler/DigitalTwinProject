```python
import pandas as pd


# read the heartrate_by_seconds data
df_hr = pd.read_csv("../FitbitData/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteSleep_merged.csv")
df_hr['datetime'] = pd.to_datetime(df_hr['date'])  # 假设时间列叫 Time
df_hr = df_hr.set_index('datetime')
```

```python

# group by userId and resample by "hour"
df_hourly = (
    df_hr
    .groupby('Id')
    .resample('h')   # 每小时聚合
    .agg({
        'value': ['mean', 'sum', 'max', 'min']
    })
    .reset_index()
)

# 展平列名
df_hourly.columns = ['Id', 'datetime', 'Value_mean', 'Value_sum', 'Value_max', 'Value_min']

print(df_hourly.head())
```

```python
df_hourly.to_csv("../FitbitData/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlySleep_merged_1.0.csv", index=False)
```
