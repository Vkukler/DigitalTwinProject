## Dealing with irregular timestamps in Heart Rate 
Goal: convert the date with irregular timestamp into a regular regular 10-second time series.
1. upsample the data to every second
2. apply forward-fill to fill the missing timestamps
3. then group into 10-second bins and take the mean.

```python
import pandas as pd

# Read your data (assuming saved as heart_rate.csv)
df = pd.read_csv("/Users/lingwenxu/Desktop/DigitalTwinGroupProject/src/data/user_5577150313_heartrate_seconds_merged.csv")

# Convert Time to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Set Time as index
df = df.set_index('Time')

# Step 1: Upsample to 1-second frequency and forward fill missing values
df_1s = df.resample('1s').ffill()

# Step 2: Downsample to 10-second bins and take the mean
df_10s = df_1s.resample('10s').mean()

# Step 3: Fill Id (keep the first Id in each 10s bin)
df_10s['Id'] = df_1s['Id'].resample('10s').first()

# Reset index if needed
df_10s = df_10s.reset_index()

print(df_10s.head(20))
```

```python
df_10s.to_csv('../src/data/user_5577150313_heartrate_seconds_preprocessed.csv', index=False)
```
